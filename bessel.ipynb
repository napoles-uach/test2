```
from numpy import mgrid
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
from scipy import special
x,y=mgrid[-25:25:500j,-25:25:500j]
r=np.sqrt(x**2+y**2)
s=special.j1(r)*25
fig = plt.figure(figsize=(10, 8))
plt.imshow(s,extent=[-25,25,-25,25])
plt.colorbar()
```
